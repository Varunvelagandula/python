```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic daily data for one year
dates = pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")
np.random.seed(42)
prices = 100 + np.cumsum(np.random.randn(len(dates)))  # Random walk

df = pd.DataFrame({"Date": dates, "Close": prices})

# Key event dates
events = {
    "EV Launch": "2024-01-15",
    "Q1 Earnings": "2024-03-30",
    "Q2 Earnings": "2024-06-30",
    "Merger Announcement": "2024-08-10",
    "Q3 Earnings": "2024-09-30",
    "Supply Chain Disruption": "2024-10-20",
    "Q4 Earnings": "2024-12-30"
}

plt.figure(figsize=(14, 6))
plt.plot(df["Date"], df["Close"])

# Add gridlines
plt.grid(True)

# Annotate events
for label, date in events.items():
    x = pd.to_datetime(date)
    y = df.loc[df["Date"] == x, "Close"].values[0]
    plt.scatter(x, y)
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 10), ha="center")

plt.title("Global Motors Inc.  Stock Price Trend with Key Events (Annual Report)")
plt.xlabel("Date")
plt.ylabel("Closing Price in USD")

plt.tight_layout()
plt.show()
```
